```python
import random
import time

# Base de datos de vecinos reales
vecinos_reales = [
    {"nombre": "Luis", "mascota": "gato", "piso": 3},
    {"nombre": "Ana", "mascota": "perro", "piso": 5},
    {"nombre": "Carlos", "mascota": "hamster", "piso": 2},
    {"nombre": "Mar√≠a", "mascota": "ninguna", "piso": 4}
]

# Frases de monstruos intentando imitar
respuestas_monstruo = [
    "Eeeeh... s√≠‚Ä¶ yo tengo un... animal terrestre dom√©stico... creo...",
    "Mi piso es‚Ä¶ el n√∫mero‚Ä¶ *ruido raro*‚Ä¶ 99?",
    "Yo soy totalmente humano. ¬øPuedo entrar?",
    "Tengo una mascota llamada‚Ä¶ *gru√±ido*‚Ä¶ digo‚Ä¶ Juan."
]

def generar_persona():
    """Decide si llega un vecino real o un monstruo."""
    es_monstruo = random.choice([True, False])

    if not es_monstruo:
        vecino = random.choice(vecinos_reales)
        return {"tipo": "vecino", **vecino}

    else:
        # El monstruo intenta imitar a un vecino random
        vecino = random.choice(vecinos_reales)
        return {
            "tipo": "monstruo",
            "nombre": vecino["nombre"],  # intenta copiar el nombre
            "mascota": random.choice([vecino["mascota"], "???", "drag√≥n"]),
            "piso": random.choice([vecino["piso"], 13, 0, None]),
            "respuesta_rara": random.choice(respuestas_monstruo)
        }

def jugar_ronda():
    persona = generar_persona()

    print("\nüîî Tocan la puerta‚Ä¶")
    time.sleep(1)

    print(f"üë§ ‚Äî Hola, soy {persona['nombre']}. ¬øPuedo entrar?")
    time.sleep(1)

    # Pregunta 1
    print("\n‚ùì Pregunta: ¬øTienes mascota?")
    if persona["tipo"] == "vecino":
        print(f"üë§ ‚Äî S√≠, tengo {persona['mascota']}.")
    else:
        print(f"üë§ ‚Äî {persona['respuesta_rara']}")
    time.sleep(1)

    # Pregunta 2
    print("\n‚ùì Pregunta: ¬øEn qu√© piso vives?")
    if persona["tipo"] == "vecino":
        print(f"üë§ ‚Äî Vivo en el piso {persona['piso']}.")
    else:
        print(f"üë§ ‚Äî Piso... {persona['piso']}.")
    time.sleep(1)

    # Decisi√≥n del jugador
    decision = input("\nüëâ ¬øLo dejas pasar? (si/no): ").lower()

    if persona["tipo"] == "vecino" and decision == "si":
        print("‚úî Correcto: era un vecino real.")
        return True
    elif persona["tipo"] == "monstruo" and decision == "no":
        print("‚úî Bien hecho: rechazaste a un monstruo.")
        return True
    elif persona["tipo"] == "vecino" and decision == "no":
        print("‚ùå Era un vecino real, lo rechazaste por error.")
        return False
    else:
        print("‚ùå ¬°ERA UN MONSTRUO! Lo dejaste entrar.")
        return False

# Juego principal
def iniciar_juego():
    print("=== ESE NO ES MI VECINO ‚Äî PROTOTIPO TEXTO ===")
    print("Identifica si es un vecino real o un monstruo.\n")

    puntuacion = 0

    for ronda in range(1, 6):
        print(f"\n--- RONDA {ronda} ---")
        if jugar_ronda():
            puntuacion += 1
        else:
            puntuacion -= 1

    print("\n=== RESULTADO FINAL ===")
    print(f"Tu puntuaci√≥n fue: {puntuacion}")

    if puntuacion > 3:
        print("üèÜ ¬°Eres un guardi√°n experto del edificio!")
    elif puntuacion >= 0:
        print("üëç No lo hiciste mal.")
    else:
        print("üëÄ Muchos monstruos se infiltraron... cuidado.")

iniciar_juego()
```
